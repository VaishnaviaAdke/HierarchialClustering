```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(travel_data, method="ward"))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,
                                 linkage='ward')
Y_pred = cluster.fit_predict(travel_data)
```

```python
travel_data["Clusters"]=Y_pred
travel_data.head(30)
```

```python
travel_data["Clusters"].value_counts()
```

```python
cluster_0_travel_data=travel_data[travel_data["Clusters"]==0]
print(cluster_0_travel_data.shape)
cluster_0_travel_data.head()
```

```python
cluster_1_travel_data=travel_data[travel_data["Clusters"]==1]
print(cluster_1_travel_data.shape)
cluster_1_travel_data.head()
```

```python
cluster_2_travel_data=travel_data[travel_data["Clusters"]==2]
print(cluster_2_travel_data.shape)
cluster_2_travel_data.head()
```

```python
travel_data["Clusters"]=travel_data.Clusters.map({0:"Negative",1:"Neutral",2:"Positive"})
travel_data.head()
```
